```python
#income ranges are categories three we will use bar plot
```

```python
income_info = data.groupby("Income_ranges").size()
```

```python
income_info
```

```python
income_info.index
```

```python
income_info.values
```

```python
color = (0.8, # redness
         0.4, # greenness
         0.2, # blueness
         )
sms.barplot( x = income_info.index,  y = income_info.values, color = color, palette = "Set1")
```

```python

sms.barplot( x = income_info.index,  y = income_info.values, palette = "pastel")
```

```python
data["Income_ranges"] = pd.cut(data["Income"],bins = [min(data["Income"]),20000,40000,60000,80000,100000,max(data["Income"])])
income_info = data.groupby("Income_ranges").size()
sms.barplot( x = income_info.index,  y = income_info.values, palette = "pastel")
```

```python
import matplotlib.pyplot as plt
data["Income_ranges"] = pd.cut(data["Income"],bins = [min(data["Income"]),20000,40000,60000,80000,100000,max(data["Income"])])
income_info = data.groupby("Income_ranges").size()
sms.barplot( x = income_info.index,  y = income_info.values, palette = "pastel")
plt.xticks(rotation = 45)
```
